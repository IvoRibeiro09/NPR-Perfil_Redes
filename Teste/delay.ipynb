O documento "NPR2324_DelayTolerantNetworks.pdf" aborda as Redes Tolerantes a Atrasos (DTNs). Aqui está um resumo detalhado com os pontos chave:

### 1. Introdução às DTNs
- **Origem e Conceito:**
  - Derivadas das MANET (Mobile Ad Hoc Networks).
  - Redes sem conectividade permanente entre todos os nós.
  - Utilizam o paradigma "store-carry-and-forward" para encaminhamento de pacotes.
  - Enfrentam o desafio da mobilidade dos nós.

### 2. Aplicações das DTNs
- **Conectividade em Regiões de Desenvolvimento:**
  - Estudantes em áreas sem infraestrutura de rede.
  - Utilização de autocarros para transportar pacotes para pontos de acesso à Internet.
- **Regiões Disconectadas:**
  - Exemplo: Parques naturais protegidos, áreas habitadas por esquimós, minas.
- **Monitorização Ambiental e Vida Animal:**
  - Redes de sensores para monitorização de animais, como o projeto ZebraNet.
- **Aplicações Militares:**
  - Uso de "Throwboxes" para melhorar a comunicação em combate.
- **Comunicações Submarinas e Redes Veiculares:**
  - Exemplo: Pocket switched networks para redes urbanas ou sociais.

### 3. Características das DTNs
- **Estrutura e Funcionamento:**
  - Compostas por nós interligados por links unidirecionais.
  - Links podem se estabelecer ou desfazer por várias razões (obstruções, falha de energia, mobilidade).
  - Mensagens podem ser armazenadas por longos períodos até um "contacto" estar disponível.
  - Conceito de "contacto" como uma oportunidade de enviar dados.
- **Tempo de Transferência de Mensagens:**
  - Inclui tempo de espera por contacto, tempo na fila, tempo de transmissão e tempo de propagação.

### 4. Principais Desafios das DTNs
- **Escalonamento dos Contactos:**
  - Variabilidade no tempo de espera por contacto.
  - Contactos podem ser previsíveis, aproximados, implícitos ou aleatórios.
- **Capacidade dos Contactos:**
  - Quantidade de dados transferível em cada contacto depende da tecnologia e duração do contacto.
- **Espaço de Armazenamento e Capacidade de Processamento:**
  - Mensagens podem exigir grande capacidade de armazenamento temporário.
  - Dispositivos pequenos podem ter limitações de CPU e memória.
- **Energia:**
  - Limitações de energia dos nós móveis.
  - Estratégias de encaminhamento podem otimizar o consumo energético.

### 5. Critérios de Avaliação das DTNs
- **Taxa de Entrega:**
  - Percentagem de mensagens geradas que são efetivamente entregues.
- **Latência:**
  - Tempo desde que a mensagem é gerada até ser recebida.
- **Número de Transmissões:**
  - Medida da capacidade dos contactos consumida pelas estratégias de encaminhamento.

### 6. Encaminhamento em DTNs
- **Desafios e Abordagens:**
  - Protocolos convencionais não são aplicáveis devido à falta de conectividade fim a fim.
  - Várias estratégias baseadas em replicação e conhecimento da rede.
- **Estratégias Específicas:**
  - **Encaminhamento Epidémico:**
    - Espalha pacotes como um vírus, inundando a rede com cópias.
    - Consome muitos recursos, requer controle de flooding.
  - **Contacto Direto:**
    - O emissor espera até encontrar o receptor, uma única cópia na rede.
  - **Reenvio em Dois Saltos:**
    - O emissor copia a mensagem para N nós; esses nós entregam ao destinatário.
  - **Reenvio em Árvore:**
    - Mensagens são copiadas com instruções sobre quantas cópias fazer e quantos saltos permitir.
  - **Baseado em Localização e Gradiente:**
    - Encaminhamento na direção esperada do destinatário ou baseado em métricas associadas aos nós.
